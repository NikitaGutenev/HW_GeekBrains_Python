```python
from sympy import *
import math
import matplotlib as plt
x = Symbol('x')
fx = -12*x**4 - 18*x**3 + 5*x**2 +10*x - 30
fx
```

```python
#корни
solveset(fx,x) 
```

```python
# интервалы где функция возрастает и убывает
a = diff(fx)
solve(a>0)
solve(a<0)
```

```python
solveset(plot(fx))
```

```python
#вершины функции
solve(a)
```

```python
#промежуток где f>0
solve(fx>0)
```
